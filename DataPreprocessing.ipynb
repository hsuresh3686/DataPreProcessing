If Skewness is Greater than -0.5 and lesser than 0.5 then using Mean to fill -0.5 < skew < 0.5
Out of the range (-0.5 > skew > 0.5) filling with Median

```python
for col in quan:
    skew=dataset[col].skew()
    print(col,"\t",skew)
```

```python
from sklearn.impute import SimpleImputer
mean_imputer = SimpleImputer(strategy='mean')
median_imputer = SimpleImputer(strategy='median')

for col in quan:
    skew=dataset[col].skew()
    if -0.5 < skew < 0.5:
        print("Mean Filled : \t", col,"\t",skew)
        dataset[[col]] = mean_imputer.fit_transform(dataset[[col]])
    else:
        print("Median Filled : \t", col,"\t",skew)
        dataset[[col]] = median_imputer.fit_transform(dataset[[col]])
    
    dataset[col] = dataset[col].round(1)

dataset['bp'] = dataset['bp'].round(0).astype(int)
```

```python
dataset[qual].isna().sum()
```

```python
# Dropping the Null values if less than 5 in Categorical Columns
for col in qual:
    if (dataset[col].isna().sum()) < 5:
        dataset = dataset.dropna(subset=[col])
```

```python
dataset[qual].isna().sum()
```

```python
frequent_imputer = SimpleImputer(strategy='most_frequent')
for col in qual:
    dataset[[col]] = frequent_imputer.fit_transform(dataset[[col]])
```
